# Review and Rating Dataset Combine

```python
df1 = pd.read_csv("2-wheeler-EV-bikewale.csv")
df2 = pd.read_csv("4-wheeler-EV-cardekho.csv")
df3 = pd.read_csv("4-wheeler-EV-carwale.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
df1 = df1.loc[:, df1.columns.intersection(['rating','Model Name'])]
df1['Type'] = '2-wheeler'
df1.rename(columns={'rating':'Rating'},inplace=True)
df1.head()
```

```python
df2 = df2.loc[:, df2.columns.intersection(['Rating','Model'])]
df2['Type'] = '4-wheeler'
df2.rename(columns={'Model':'Model Name'},inplace=True)
df2.head()
```

```python
df3 = df3.loc[:, df3.columns.intersection(['rating','model_name'])]
df3['Type'] = '4-wheeler'
df3.rename(columns={'rating':'Rating','model_name':'Model Name'},inplace=True)
df3.head()
```

```python
df4 = pd.concat([df1,df2,df3],ignore_index = True , sort=False)
df4.head()
```

```python
df4.info()
```

```python
df4.to_csv (r'combine_rating_all_vehicle.csv', index = False, header=True)
```
